```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt


file_path = 'data_vnm.xlsx'  
data = pd.read_excel(file_path)

# Convert the 'Ngay' column to datetime format for filtering
data['Ngay'] = pd.to_datetime(data['Ngay'])

# Extract the 'GiaDongCua' column and preprocess it
closing_prices = data['GiaDongCua'].values.reshape(-1, 1)

# Train Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.5, random_state=42)
data['anomaly'] = iso_forest.fit_predict(closing_prices)

# Mark anomalies: -1 indicates an anomaly
anomalies = data[data['anomaly'] == -1]

# Filter data for during COVID 
during_covid = data[(data['Ngay'] >= '2019-12-01') & (data['Ngay'] <= '2021-12-31')]

print('bang during covid')
# print(during_covid[during_covid])
# Plotting the data during COVID-19
plt.figure(figsize=(10, 6))
plt.plot(during_covid['Ngay'], during_covid['GiaDongCua'], label='GiaDongCua', color='blue')
plt.scatter(during_covid[during_covid['anomaly'] == -1]['Ngay'], during_covid[during_covid['anomaly'] == -1]['GiaDongCua'], color='red', label='Anomaly', marker='x')
plt.title('During-COVID: Anomaly Detection on GiaDongCua')
plt.xlabel('Ngay')
plt.ylabel('GiaDongCua')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
